```python
from random import shuffle

```

```python
def check_rules(update, rules):
    update_dict = {}
    for idx, val in enumerate(update):
        update_dict[val] = idx
    
    is_rule_held = True
    for rule in rules:
        before, after = rule.split('|')
        if before in update_dict and after in update_dict:
            is_rule_held = update_dict[before] < update_dict[after]
            if is_rule_held:
                continue
            else:
                #print(rule, 'broken')
                break

            
            
    return is_rule_held            
```

```python
from tqdm import tqdm

corrected_updates = []
for bad_update in tqdm(incorrect_updates):
    is_incorrect = True
    while is_incorrect:
        shuffle(bad_update)
        is_incorrect = False if check_rules(bad_update, rules) else True
    
    corrected_updates.append(bad_update)
```
