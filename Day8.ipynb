```python
# find the antinodes of the antennae
# given 2 antennae of the same type, 2 antinodes are produced. 
# an antinode occurs at any point that is perfectly in line with two antennas of the same frequency - 
# but only when one of the antennas is twice as far away as the other. 
# This means that for any pair of antennas with the same frequency, there are two antinodes, one on either side of them.

# goal: find the number of unique positions in the map that contain antinodes

# some rules:
# antinodes can occur at locations that contain antennas
# multiple antinodes can occur at the same spot I think? Either way it is unique locations.
```

```python
# strategy: treat this as a grid. Find all xy points of all unique node types
# for each node type, iterate through all pairs of nodes. 
# calculate the slope-intercept line
# use slope intercept line to calculate anti-node position
# if position inside map, mark on grid, etc. 
```

```python
data_split = data.split('\n')
citymap = [[x for x in i] for i in data_split]
```

```python
citymap
```

```python
node_positions = {}
for i in range(len(citymap)):
    for j in range(len(citymap[i])):
        if citymap[i][j] != '.':
            antenna = citymap[i][j]
            if antenna in node_positions:
                node_positions[antenna].append((i,j))
            else:
                node_positions[antenna] = [(i,j)]
```

```python
for antenna in node_positions.keys():
    antenna_positions = node_positions[antenna]
    # start a loop here 
    break
```

```python
for pos1 in antenna_positions:
    for pos2 in antenna_positions:
        if pos1 != pos2:
            break
    break
```

```python
def calculate_slope(pos1, pos2):
    return (pos2[1] - pos1[1]) /(pos2[0] - pos1[0]) 

def calculate_b(pos, slope):
    return pos[1] - (slope * pos[0])

def calculate_distance(pos1, pos2):
    # return the distance between the two points, on the grid
    #return ((pos2[0] - pos1[0])**2 + (pos2[1] - pos1[0])**2)**.5
    return abs(pos2[0] - pos1[0]) + abs(pos2[1] - pos1[1])
```

```python

# Fixed working solution. Instead of calculating the slope and intercept, I just used the difference in coordinates. 
# The problem was too many problems with slope and intercept if not integers. 
working_positions = []
for antenna in node_positions.keys():
    antenna_positions = node_positions[antenna]
    for pos1 in antenna_positions:
        for pos2 in antenna_positions:
            if pos1 != pos2:
                row_diff = pos2[0] - pos1[0]
                col_diff = pos2[1] - pos1[1]

                new_pos = (pos1[0] - row_diff, pos1[1] - col_diff)


                if 0 <= new_pos[0] < len(citymap) and 0 <= new_pos[1] < len(citymap[0]):
                            
                    if new_pos not in working_positions:
                        working_positions.append(new_pos)
                                

len(working_positions)
```

```python

# Fixed working solution. Instead of calculating the slope and intercept, I just used the difference in coordinates. 
# The problem was too many problems with slope and intercept if not integers. 

# lazy solution: use the row and col diff calculation, repeat N times, and only retain solutions inside the city
working_positions = []
for antenna in node_positions.keys():
    antenna_positions = node_positions[antenna]
    for pos1 in antenna_positions:
        for pos2 in antenna_positions:
            if pos1 != pos2:
                row_diff = pos2[0] - pos1[0]
                col_diff = pos2[1] - pos1[1]
                for i in range(0,1000):
                    new_pos = (pos1[0] - i*row_diff, pos1[1] - i*col_diff)


                    if 0 <= new_pos[0] < len(citymap) and 0 <= new_pos[1] < len(citymap[0]):
                                
                        if new_pos not in working_positions:
                            working_positions.append(new_pos)
                                

len(working_positions)
```
