```python
# basic idea here can solve it I think...

# first find the index of the guard (row,col) 
# next, determine the orientation [^,<,v,>]
# based on the orientation, move forward until an obstacle is in front of you #
# when obstacle is reached, rotate the orientation 90 degrees. 
# continue until the guard reaches a boundary point
# count all of the steps taken before.
```

```python
data_split = data.split('\n')
guard_path = [[x for x in y] for y in data_split]
```

```python
guard_orientations = ['^','>','v','<']
obstacles = ['#']

def get_guard_position_and_orientation(guard_path, guard_orientations):
    for row in range(len(guard_path)):
        for col in range(len(guard_path[0])):
            value = guard_path[row][col]
            if value in guard_orientations:
                guard_orientation = [i for i in guard_orientations if i == value][0]
                guard_position = (row,col)
                return guard_position, guard_orientation


def get_new_orientation(g):

    if g == '^':
        return '>'
    elif g == '>':
        return 'v'
    elif g == 'v':
        return '<'
    elif g == '<':
        return '^'

def move_guard(guard_path, guard_orientation, guard_position,obstacles):
    is_inside = True
    guard_x = guard_position[0]
    guard_y = guard_position[1]
    if guard_orientation == '^':
        next_x = guard_x - 1
        next_y = guard_y 
    
    if guard_orientation == 'v':
        next_x = guard_x + 1
        next_y = guard_y 

    if guard_orientation == '<':
        next_x = guard_x 
        next_y = guard_y - 1
    
    if guard_orientation == '>':
        next_x = guard_x 
        next_y = guard_y + 1

    if next_x not in list(range(0,len(guard_path))):
        is_inside = False
        next_x = guard_x
        next_y = guard_y 
        return (next_x, next_y), guard_orientation, is_inside

    if next_y not in list(range(0,len(guard_path[0]))):
        is_inside = False
        next_x = guard_x
        next_y = guard_y 
        return (next_x, next_y), guard_orientation, is_inside

    if guard_path[next_x][next_y] in obstacles:
        # rotate instead of continuing 
        next_x = guard_x
        next_y = guard_y 
        guard_orientation = get_new_orientation(guard_orientation)
    

    return (next_x, next_y), guard_orientation, is_inside
```

```python
is_inside = True
guard_position, guard_orientation = get_guard_position_and_orientation(guard_path, guard_orientations)
guard_path[guard_position[0]][guard_position[1]] = 'X'
while is_inside: 

    guard_position, guard_orientation, is_inside = move_guard(guard_path, guard_orientation, guard_position,obstacles)
    guard_path[guard_position[0]][guard_position[1]] = 'X'
    
```

```python
from tqdm import tqdm
obstacles = ['#','O']

obstacle_locations = []
num_stuck = 0
for guard_path, guard_path_init in tqdm(zip(possible_new_paths, original_guard_paths)):
    is_inside = True

    guard_position, guard_orientation = get_guard_position_and_orientation(guard_path_init, guard_orientations)
    is_stuck = False
    seen_position_orientations = []
    while is_inside: 
        seen_str = f"{guard_orientation}-{guard_position[0]}-{guard_position[1]}"        
        guard_position, guard_orientation, is_inside = move_guard(guard_path, guard_orientation, guard_position,obstacles)
        if seen_str in seen_position_orientations:
            is_stuck = True 
            if is_inside:
                obstacle_locations.append(seen_str)
                num_stuck += 1
            break

        seen_position_orientations.append(seen_str)
```
