```python
# Part 2: problem dampener will help increase number of safe by allowing you to remove the bad value.

# If you can remove one bad level from the input, does the report become safe?

# Always remove the first bad input?
# Removing will never help with the problem of it being too high, unless that is the first or last value
# It MAY help if there are repeat values assuming only 1
# It MAY help with increasing if it happens once. 

# adversarial case: 1,7,5,4
# Remove the first level to make always decrease 
```

```python
# pseudo code
# for level in report 
# get prior, current, and leading reading
# can determine if safe using all 3 of these

# values that matter
# delta between current and prior
# delta between current and next
# delta between next and next next 
```

```python
def is_report_safe(report):
    prior_is_increasing = False
    for i in range(1,len(report)):
        prior_reading = report[i-1]
        current_reading = report[i]
        delta = current_reading - prior_reading
        if delta > 0:
            is_increasing = True
        else:
            is_increasing = False
#        print(delta, current_reading, prior_reading, is_increasing, prior_is_increasing, i)

        if i > 1:
            if is_increasing != prior_is_increasing:
                # update num unsafe, stop this report 
                return False
        if abs(delta) > 3:
            return False
            # update num unsafe, stop this report
        if abs(delta) == 0:
            return False
            # update num unsafe, stop this report

        prior_is_increasing = is_increasing
        if i == len(report) - 1:
            return True
```

```python
reports_int_test = [[7,6,4,2,1],[1,2,7,8,9], [8,6,4,4,1], [1,7,6,5,4]]
num_safe = 0
for report in reports_int:
    # brute force soln. Just remove one level every time. 

    if is_report_safe(report):
        num_safe += 1

    else:
        for i in range(len(report)):
            report_level_removed = report[:i] + report[i+1:]
            if is_report_safe(report_level_removed):
                num_safe += 1
                break
```
