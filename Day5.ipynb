```python
# copy this above. Swap the indices of the rules when broken. Make recursive, only stop if all rules are done. 

f = open("day5data.txt", "r")
data = f.read()

rules, updates = data.split("\n\n")

rules = rules.split("\n")
updates = updates.split("\n")
updates = [x.split(',') for x in updates]


def correct_any_issues(update, rules):


    # create dict for better tracking 
    update_dict = {}
    for idx, val in enumerate(update):
        update_dict[val] = idx

    # loop through rules. Check if any violations exist and swap the positions if they do
    is_rule_held = False
    while not is_rule_held:
        is_rule_held = True
        for rule in rules:
            before, after = rule.split('|')
            if before in update_dict and after in update_dict:
                is_rule_held = update_dict[before] < update_dict[after]
                if is_rule_held:
                    continue
                else:
                    # swap 
                    before_idx = update_dict[before]
                    after_idx = update_dict[after]
                    # fix the update

                    update[before_idx], update[after_idx] = update[after_idx], update[before_idx]
                    # now recreate updates dict
                    update_dict = {}
                    for idx, val in enumerate(update):
                        update_dict[val] = idx
                

    return update
```
